```python
def lang_confidence_score(word_counts, language_words_with_frequency):
    """ 
    word_counts: słownik {slowo: liczba_wystapien_w_tekscie}
    language_words_with_frequency: słownik {slowo: czestotliwosc_w_jezyku} (dla top k słów)
    """
    score = 0
    total_words_in_text = sum(word_counts.values())

    if total_words_in_text == 0:
        return 0

    # Iterujemy po słowach z badanego tekstu:
    for word, count in word_counts.items():
        # Sprawdzamy, czy słowo jest w zestawie top-k słów języka
        if word in language_words_with_frequency:            
            # Ważenie "popularnością" w języku
            score += count * language_words_with_frequency[word]

    # Normalizacja wyniku:
    normalized_score = score / total_words_in_text
    
    return normalized_score
```
